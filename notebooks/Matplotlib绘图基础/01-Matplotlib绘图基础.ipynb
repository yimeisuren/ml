**案例1: 单x --- 单y**

````python
dataX = df['date']
dataY1 = df['unrate']
plt.plot(dataX, dataY1)
plt.show()
````

添加图表信息
+ xlabel
+ ylabel
+ title

````python
dataX = df['date']
dataY1 = df['unrate']
plt.plot(dataX, dataY1)
plt.xlabel('Year')
plt.ylabel('Unrate')
plt.title('失业率信息')
plt.xticks(rotation=45)
plt.show()
````

**案例2: 单x --- 多y**

````python
dataX = df['date']
dataY1 = df['unrate']
dataY2 = df['unrate_16_to_17']

plt.plot(dataX, dataY1, color='darkred', label='unrate')
plt.plot(dataX, dataY2, color='darkblue', label='unrate_16_to_17')

# 添加图例
plt.legend(loc='best')
plt.show()
````

使用**列表 + 循环**进行改进

````python
dataX = df['date']
dataY1 = df['unrate']
dataY2 = df['unrate_16_to_17']

# 准备好参数列表
X = [dataX]
Y = [dataY1, dataY2]
#RGB三元组需要除以255
colors = ['darkred', (0 / 255, 0 / 255, 255 / 255)]
labels = ['unrate', 'unrate_16_to_17']
for i in range(len(Y)):
    plt.plot(X[i % len(X)], Y[i % len(Y)], color=colors[i % len(colors)], label=labels[i % len(labels)])

plt.legend(loc='best')
plt.show()
````

**案例3: 子图**

(row, columns, index)

例如(2, 2, 1)表示下方序号为1的区域
```
1, 2
3, 4
```

````python
figure = plt.figure(
    figsize=[6, 6]  # 指定子图的长和宽
)  # 添加画图域
subplot_1 = figure.add_subplot(2, 2, 1)
subplot_2 = figure.add_subplot(2, 2, 2)
subplot_3 = figure.add_subplot(2, 2, 3)
subplot_4 = figure.add_subplot(2, 2, 4)
plt.show()
````

````python
dataX = df['date']
dataY1 = df['unrate']
dataY2 = df['unrate_16_to_17']

figure = plt.figure(
    figsize=[6, 6],  # 指定子图的长和宽
)  # 添加画图域
subplot_1 = figure.add_subplot(3, 2, 1)
subplot_2 = figure.add_subplot(3, 2, 2)
subplot_3 = figure.add_subplot(3, 2, 3)
subplot_4 = figure.add_subplot(3, 2, 4)
subplot_5 = figure.add_subplot(3, 2, 5)
subplot_6 = figure.add_subplot(3, 2, 6)

# 绘图
subplot_1.plot(dataX, dataY1)
subplot_1.set_xlabel('Year')
subplot_1.tick_params(bottom='on',
                      top='off',
                      left='off',
                      right='off')
subplot_1.set_xticklabels(dataX.apply(lambda date: date.year),
                          rotation=90)

subplot_2.bar(dataX, dataY2)  #竖着画

subplot_3.barh(dataX, dataY2)  #横着画

subplot_4.scatter(dataX, dataY2)

subplot_5.hist(dataX,
               bins=5)

selectedColumns = ['unrate', 'unrate_16_to_17']
subplot_6.boxplot(df[selectedColumns])  # 盒图
subplot_6.set_xticklabels(selectedColumns, rotation=90)
subplot_6.set_ylim(0, 50)  #设置y的上下限
# 显示
plt.show()
````
