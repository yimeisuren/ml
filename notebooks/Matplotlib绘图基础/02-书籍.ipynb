```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

matplotlib.rcParams['font.sans-serif'] = ['SimSun', 'SimHei']
matplotlib.rcParams['mathtext.default'] = 'tt'
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams['pgf.texsystem'] = 'xelatex'


def f1(i):
    return np.exp(-i) * np.cos(2 * np.pi * i)


def f2(i):
    return 1 / 100 * i ** 2


x = np.arange(0, 10, 1)
y1 = np.array([f1(i) for i in x])
y2 = np.array([f2(i) for i in x])

xmin = x.min() - 1
xmax = x.max() + 1
ymin = min(y1.min(), y2.min()) - 1
ymax = max(y1.max(), y2.max()) + 1
```

# x轴的默认数据: 0,1,2,...

```python
plt.plot(y1)
plt.ylabel('some number')
plt.show()
```

# plot(): 在指定绘图区域进行绘图

label + legend(): 添加图例

```python
plt.plot(x, y1, 'ro', label='Dot')
plt.axis([xmin, xmax, ymin, ymax])

plt.plot(x, y1, 'b-', label='Line 1')

plt.plot(x, y2, 'y1--', label='Line 2')

plt.legend()
plt.show()
```

# figure(i:int)指定当前图形, 最终绘图顺序和i无关

```python
# 第1张图
plt.figure(1)

# 第一张图下的1号子图
plt.subplot(3, 1, 1)
plt.plot(x, y1, 'b-')

# 第一张图下的2号子图
plt.subplot(3, 1, 2)
plt.plot(x, y2, 'y1--')

# 第2张图
plt.figure(2)
plt.plot(x, y1, 'y1-')

# 指定当前图形为第1张图
plt.figure(1)
plt.subplot(3, 1, 3)
plt.plot(x, y2, 'b-')
plt.title('Easy as 1, 2, 3')

plt.show()
```

# 文本标注

```python
µ, σ = 100, 15
data = µ + σ * np.random.rand(10000)

n, bins, patches = plt.hist(data, 50, facecolor='g', alpha=0.75)

plt.xlabel('XLabel', fontsize=14, color='red')
plt.ylabel('YLabel')
plt.title('Title')
plt.text(60, 0.025, fr'$µ={µ}, σ={σ}$')
plt.axis([40, 160, 0, 0.03])

plt.show()
```

# subplot2grid()将每一个标准绘图区域看作是一个单元格

colspan和rowspan类似于excel中的合并单元格

```python
ax1 = plt.subplot2grid((2, 2), (0, 0), colspan=2)
plt.plot(x, y1, 'b-')
ax2 = plt.subplot2grid((2, 2), (1, 0))
plt.plot(x, y2, 'r-')
ax3 = plt.subplot2grid((2, 2), (1, 1))
plt.plot(x, y1, 'r-')

plt.show()
```

# gridspec提供数组切片的方式来合并单元格

```python
import matplotlib.gridspec as gridspec

# 划分标准绘图区域 3*3
gs = gridspec.GridSpec(3, 3)

# 子图1
plt.subplot(gs[0, 0:])
plt.plot(x, y1, 'b-')
plt.title('ax1')

# 子图2
plt.subplot(gs[1, 0:-1])
plt.plot(x, y1, 'r-')
plt.title('ax2')

# 子图3
plt.subplot(gs[1:, -1])
plt.plot(x, y2, 'b-')
plt.title('ax3')

# 子图4
plt.subplot(gs[-1, 0])
plt.plot(x, y2, 'r-')
plt.title('ax4')

# 子图5
plt.subplot(gs[-1, 1])
plt.plot(x, y2, 'b--')
plt.title('ax5')

plt.show()
```

# tight_layout(): 调整子图间的间隔

```python
import matplotlib.gridspec as gridspec

gs = gridspec.GridSpec(3, 3)

plt.subplot(gs[0, 0:])
plt.plot(x, y1, 'b-')
plt.title('ax1')

plt.subplot(gs[1, 0:-1])
plt.plot(x, y1, 'r-')
plt.title('ax2')

plt.subplot(gs[1:, -1])
plt.plot(x, y2, 'b-')
plt.title('ax3')

plt.subplot(gs[-1, 0])
plt.plot(x, y2, 'r-')
plt.title('ax4')

plt.subplot(gs[-1, 1])
plt.plot(x, y2, 'b--')
plt.title('ax5')

# 紧密布局: 解决子图之间文本重合的问题
plt.tight_layout()
plt.show()
```

# 添加注释内容

+ xy: 注释位置
+ xytext: 文本位置

```python
data = np.linspace(-np.pi, 2 * np.pi, 1000)
plt.plot(np.pi, 0, 'ro')

plt.plot(data, np.sin(data), 'b-')
plt.annotate('$y = α * sin(x) + β * cos(x)$', xy=(np.pi, 0), xytext=(2 * np.pi - 1, 1),
             arrowprops=dict(
                 arrowstyle='-|>',
                 connectionstyle='arc3',
                 facecolor='black')
             )
plt.title("中文标题")

plt.show()
```
